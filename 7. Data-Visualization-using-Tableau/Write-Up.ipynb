## Summary

“Baseball Data” was chosen as the dataset for Tableau’s final project. It contains the following information on 1157 baseball players:

-	The handedness (right, left or both handed)
-	Height (in inches)
-	Weight (in pounds)
-	Batting average
-	Home runs

In this project, through data analysis and visualizations we shall try to understand the relationship between a players' physical attributes and performance.

## Design

Data Exploration:
I started with minor editing of the data in the downloaded csv file “baseball_data.csv”:

-	Except for the “HR” column, changing all the column names’ first letter to capital.
-	Replacing the ‘B’, ‘L’ & ‘R’ initials under the “Handedness” column to “Both”, “Left” & “Right” respectively.

(Note: All of the above could have been done in the Tableau itself, but for some reason, the edited names in the project weren’t recognized once the project was closed and re-opened again, there were duplication of the column names from the original file.)

With my idea on baseball a bit limited, I did a quick read online to gather basic knowledge on batting average and home runs.
Going through the data in the file, I found that out of 1157 data:

-	20 players have 0 for home run where as their batting average is more than that.
-	Also, when I was cross-checking a couple of player’s profile I came across the player Tom Brown who has a 0 for both home run and batting average but his Wikipedia page shows 0.147 and 1 for batting average and home run respectively.

The above two instances led to me believe that there’s a good chance we might be missing some crucial data on the players.

Data Analysis & Visualization:

For analysis purposes, we have two separate attributes to consider:

- Physical attribute: Height, Weight, Handedness
- Performance attribute: Batting Average, Home Run

We compare each physical component to a separate performance parameter, like height vs. batting average, weight vs. home run etc., to give us a better look at what aspect has the most influence.

Next, we mix up the physical attributes to further analyze the performance.

### FINDINGS:

Here's a quick summary on the findings based on handedness, height and weight:

- **HANDEDNESS**: Most of the players are right handed (at 737) followed by left handed (at 316) and both handed (at 104) players.

    - Both handed players have the **highest mean batting average at 0.21** but have an average home run of 32.14.
    - Left handed players have the **highest average home run at 56.15** but have a mean batting average of 0.20.
    - Right handed players have a mean batting average of 0.18 and average home run of 42.60.
    - In short:
    
        - For mean batting average: Right < Left < Both
        - For average home run: Both < Right < Left

- **HEIGHT**: Players are divided into three groups for height:
    - 65 to 71 inches
    - 72 to 74 inches
    - 75 to 80 inches
    
    Most of the players are in the 72 to 74 group (at 606) followed by group 65 to 71 (at 316) and 75 to 80 (at 235).

    - Group 65 to 71 has the **highest mean batting average at 0.22** but have an average home run of 41.10.
    - Group 72 to 74 has the **highest average home run at 50.67** but have a mean batting average of 0.20.
    - Group 75 to 80 has a mean batting average of 0.12 and average home run of 37.39.
    - In short:
    
        - For mean batting average: (75 to 80) < (72 to 74) < (65 to 71)
        - For average home run: (75 to 80) < (65 to 71) < (72 to 74)

- **WEIGHT**: Players are divided into three groups of weight:
    - 140 to 175 pounds
    - 176 to 197 pounds
    - 198 to 245 pounds
    
    Most of the players are in the 176 to 197 group (at 565) followed by group 140 to 175 (at 379) and 198 to 245 (at 213).
    
    - Group 140 to 175 has the **highest mean batting average at 0.21** but have an average home run of 35.04.
    - Group 198 to 245 has the **highest average home run at 53.14** but have an mean batting average of 0.15.
    - Group 176 to 197 has a mean batting average of 0.18 and average home run of 49.35
    - In short:
    
        - For mean batting average: (198 to 245) < (176 to 197) < (140 to 175)
        - For average home run: (140 to 175) < (176 to 197) < (198 to 245)
